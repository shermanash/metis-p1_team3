```python
import pandas as pd


mta = pd.read_csv("turnstile_160109.txt")
mta = mta.head(1000)
mta
```

```python
mta['key'] = zip(mta['C/A'], mta['UNIT'], mta['SCP'], mta['STATION'])
```

```python
import datetime
mta['datetime'] = pd.to_datetime(mta.DATE + mta.TIME, format='%m/%d/%Y%H:%M:%S')

mta['day'] = pd.to_datetime(mta.DATE, format='%m/%d/%Y')


```

```python
entries_list = []


for index, row in mta.iterrows():
    if mta['TIME'].iloc[index] == '03:00:00':
        startentries = mta['ENTRIES'].iloc[index]
    if mta['TIME'].iloc[index] == '23:00:00':
        finalentries = mta['ENTRIES'].iloc[index]
        total = finalentries - startentries
        date = mta['day'].iloc[index]
        finaltuple = (date, total)
        entries_list.append(finaltuple)


        
print entries_list        
```
